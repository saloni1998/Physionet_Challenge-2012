```python
from keras.models import Model
from keras.layers import Input,Dense,Dropout,Conv1D,MaxPooling1D,Flatten,Activation,BatchNormalization
from keras.optimizers import Adam,RMSprop,SGD,Nadam,Adagrad
from keras.losses import mean_squared_error,binary_crossentropy
from keras.utils import np_utils
from keras.callbacks import ReduceLROnPlateau
```

```python
def make_model():
    inputs=Input(shape=(33,))
    l=Dense(64)(inputs)
    l=BatchNormalization()(l)
    l=Activation('relu')(l)
    
    #l=Dropout(0.5)(l)
    
    
    l=Dense(128)(l)
    l=BatchNormalization()(l)
    l=Activation('relu')(l)
    
    #l=Dropout(0.5)(l)
    
    
    l=Dense(256)(l)
    l=BatchNormalization()(l)
    l=Activation('relu')(l)
    
    l=Dropout(0.4)(l)
    
    l=Dense(256)(l)
    l=BatchNormalization()(l)
    l=Activation('relu')(l)
    
    l=Dropout(0.4)(l)
    
    l=Dense(128)(l)
    l=BatchNormalization()(l)
    l=Activation('relu')(l)
    
    l=Dropout(0.5)(l)
    
    l=Dense(64)(l)
    l=BatchNormalization()(l)
    l=Activation('relu')(l)
    
    l=Dropout(0.5)(l)
    
    
    l=Dense(2,activation='sigmoid')(l)
    outputs=l
    return inputs,outputs
    
    
    
    
```

```python
inputs,outputs=make_model()
model=Model(inputs=inputs,outputs=outputs)
model.summary()
```

```python
model.compile(loss='binary_crossentropy',
            optimizer=Adam(lr=0.005),
           metrics=['accuracy'])
```

```python
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1,
                              patience=3, min_lr=0,verbose=1)
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test), epochs=20,batch_size=128, callbacks=[reduce_lr],verbose=1)
```

```python
y_pred=model.predict(X_test)
print(confusion_matrix(np.argmax(y_test,axis=1),np.argmax(y_pred,axis=1)))
```
